```python
#  0 ==> Iris-setosa
#  1 ==> Iris-versicolor
#  2 ==> Iris-virginica
input_data=iris.drop(['output',"Id"],axis=1)
output=iris['output']
print(input_data)
print(output)
```

```python
iris.describe()
```

```python
scaler=StandardScaler()
```

```python
input_std=scaler.fit_transform(input_data)
```

```python
print(input_std)
```

```python
input_std.std()
```

```python
output.std()
```

```python
# stratify is user for split labels in in equal manner 
# first input varibules second output varibules
train_in,test_in,train_out,test_out=train_test_split(input_std,output,test_size=0.2,stratify=output)
```

```python
print(train_in.shape,train_out.shape,test_in.shape,test_out.shape)
```

```python
model_classifier=svm.SVC()
```

```python
model_classifier.fit(train_in,train_out)
```

```python
predict=model_classifier.predict(train_in)
score=accuracy_score(predict,train_out)
print(score)
```

```python
test_predict=model_classifier.predict(test_in)
score_test=accuracy_score(test_predict,test_out)
print(score_test)
```

```python
#testing the model iris-setosa
# without std prediction
specie0=np.array([4.9,3.0,1.4,0.2])
specie0=specie0.reshape(1,-1)
specie0=scaler.transform(specie0)
specie0_prediction=model_classifier.predict(specie0)
specie0_score=accuracy_score(specie0_prediction,np.array([0]))
print(specie0_score)
if specie0_prediction==0:
    print("Specie of Iris-setosa")
elif specie0_prediction==1:
    print("Specie of Iris-versicolor")
else:
    print("Specie of Iris-virginica")
```
