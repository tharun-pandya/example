```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# dividing data set into input data and output data
in_data=data_set['Message']
out_data=data_set['Category']
# checking data is divided or not
print(in_data)
print(out_data)
```

```python
# perform Label Encode operation without using LabelEncoder function
data_set.loc[data_set['Category']=="ham",'Category']=0
```

```python
out_data.loc[out_data=="spam"]=1
```

```python
# dtype:object convert into integer for better understand the model
out_data=out_data.astype(int)
```

```python
out_data.tail()
```

```python
# to convert alphabetical into numerical data
extraction=TfidfVectorizer(lowercase=True, min_df=1, stop_words="english")
# in above parameters are 
# "lowercase=True" it means all the words convert into lower case letters 
# ==> because this function consider Dog and dog two different words
# "min_df=1" it means consider the words which are greater than 1 otherwise it will not consider
# stop_words are nothing but nouns,prounouns,helping verbs etc these words are removes 
# ==> because these words are repeated several times in sentance
```

```python
in_data_extract=extraction.fit_transform(in_data)
```

```python
print(in_data_extract)
```

```python
# slpiting data into traing data and testing data
in_train,in_test,out_train,out_test=train_test_split(in_data_extract,out_data,test_size=0.3,random_state=5,stratify=out_data)
```

```python
# checking size of the data
print(in_data_extract.shape)
print(in_train.shape)
print(out_train.shape)
print(in_test.shape)
print(out_test.shape)
```

```python
# create logistic regression model
model=LogisticRegression()
```

```python
# train model with training data
model.fit(in_train,out_train)
```

```python
# checking accuracy in training data
train_pred=model.predict(in_train)
train_accu=accuracy_score(out_train,train_pred)
print("accuracy is:",train_accu)
```

```python
# checking accuracy in testing data
test_pred=model.predict(in_test)
test_accu=accuracy_score(out_test,test_pred)
print("accuracy is:",test_accu)
```
